### 1) Quantum Computing without Quantum Bits

Let's say you and your friend are spatially seperated and talking on the phone. You decide to turn your friend into a "quantum computer" by giving him the following instructions: 

1) Find two coins. Label one coin "coin A" and the other "coin B".

2) He initializes coin "B" in the "heads" state, but flips coin "A" ensuring that it undergoes several spins then catches it.

3) You tell him to check coin A's result, but not to reveal any information to you. If coin A is heads, flip coin B. However, if coin A is tails, leave coin B alone.

4) Although your friend knows the final state of the system, you don't know what it is since you can't see the final result of the coins. You ask your friend to read you the result of the coins over the phone.

Before you ask your friend for the result of the coins, you ask yourself the following questions:

- What are the possible final states you can extract from this system and what are their probabilities? Make a Python simulation of the coin flipping and extract probabilities experiementally.

- What is a possible wavefunction that could represent the state of the system? Is the wavefunction entangled?

- What is the procedure for creating this state using quantum gates? (Hint: Look up 'Bell States') Explain what each gate does in your own words.

- What could your friend's wavefunction of the system be from his point of view?

- Is your friend a qubit? (Hint: Does he satisfy the DiVincenzo's criteria?)

~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~ ~

### ***Solution***

The two possible states we can find of the system are $\ket{heads}\ket{heads}$ or $\ket{tails}\ket{tails}$

According to the code simulation below, the probability of $\ket{heads}\ket{heads}$ occuring is $0.5$ and the probability of $\ket{tails}\ket{tails}$ occuring is $0.5$. The simulation suggests that the $\ket{tails}\ket{heads}$ and the $\ket{heads}\ket{tails}$ states are impossible and cannot occur.

Since there is a $50-50$ chance of either state occuring, a wavefunction we could write to represent this probability distribution could look something like:

$$\ket{\psi} = \frac{\ket{heads}\ket{heads} + \ket{tails}\ket{tails} }{\sqrt{2}}$$

The wavefunction cannot be factored into a product of single qubit states and is therefore entangled.

If we map our 'heads' and 'tails' basis to '0' and '1' respectively, flipping coin A initially is like doing a Hadamard gate on the $\ket{0}$ state, transforming it into an equal superposition. 

Checking coin A's result and deciding to flip B if A is $\ket{1}$ and do nothing if A is $\ket{0}$ is a classical CNOT gate with the control on A and the target on B.

This procedure using quantum logic gates happens to be the exact procedure for generating a Bell State! The Haddamard flips coin A. And the CNOT decides whether not to flip coin B based on coin A's result.

Because the wavefunction is a state of knowledge and our friend is the one manipulating the coins directly based off his observations, our friend's wavefunction has complete knowledge of the system and it is always exactly one basis state -- the true state of the system. We have incomplete knowledge from our perspective, since we cannot observe the coins directly. Therefore, we must always represent our wavefunction as a superposition (with the exception of the coin preparation phase and the end measurement where we ask our friend for the answer).

Our friend does have a universal gate set! He can do single qubit rotations (Hadamards) and entanglement gates (CNOT's). He has a preparation and measurement feature since he can always reset the coins to their $heads$ state and report the final state of the coins to us over the phone. However, our friend is not scalable. Introducing more coins (qubits) would not provide the same time complexity speed ups that quantum computers guarentee. Such a computer would suffer $2^{N}$ time complexity!

```python
'''
Write coin flipping code here!
'''

import numpy as np
import matplotlib.pyplot as plt

trials = 1000

states = []

# Doing the coin flipping experiment for several trials
for k in range(trials):
    coin_A_state = np.random.randint(2) # Flipping the Coin
    coin_B_state = 0

    # Classical CNOT
    if coin_A_state == 0:
        # Leave Coin B Alone
        pass
    elif coin_A_state == 1:
        # Flip Coin B
        coin_B_state +=1
        if coin_B_state == 2:
            coin_B_state = 0

    # Recording the final result (Measurement)
    states.append(str(coin_A_state) + str(coin_B_state))


plt.hist(states)
plt.xlabel('Final State')
plt.ylabel('Number of Measurements')
plt.title('Coin Flipping Experiement')
plt.show()
```

#### 2d) The Classical Monty Hall

The Monty Hall problem is a famous probability puzzle based on a game show scenario.

🎯 The Setup:

You're on a game show, faced with three doors: behind one is a car (the prize), and behind the other two are goats.
You choose one door — say, door 1.
The host, Monty Hall, who knows where the prize is, opens one of the remaining two doors, always revealing a goat.

Now you’re given a choice:

👉 Stick with your original door

👉 Switch to the other unopened door

What should you choose and why?

(Answer: It is favorable to switch. Convince yourself by working out the probabilities)

### ***Solution***

If we pick door 1, that means that there is a 1/3 chance of that being the correct door and a 2/3 chance that it can either be doors 2 or 3. Monty reveals that door 2 is a dud, therefore our probabilities update such that door 1 has a 1/3 chance of being correct and door 3 has a 2/3 chance of being correct. Because Monty knows what the correct door is, door 3 effectively absorbed door 2's probability when it was revealed to be a dud. Therefore, the most advantageous move would be to switch from door 1 to door 3, which has a 2/3 chance of success.

#### 2e) The Quantum Monty Hall

Monty realizes that you're taking advantage of his knowledge whenever he reveals a dud door to you. He decides to not reveal anything and encode the answer in terms of a wavefunction. He hands you three qubits which form an equally weighted superposition between the three doors, forcing you to choose at random since each outcome is equally likely upon measurement. Monty claims that he's playing fair since he ensures that the 'right' answer is encoded in his wavefunction with a minus sign phase multiplied to the state with the correct door. However, he doesn't tell you which state has an appended minus sign since, if he told you, that information would obviously give away the answer. Cleverly, he knows that an additional minus sign on one of the basis states on the wavefunction doesn't affect the probability distribution, forcing you to choose an outcome at random upon measurement. You can only take one measurement shot on the wavefunction and that's it. What should you do?

Hint: The wavefunction Monty gives you could be any of these three possibilities

$$\ket{\phi_{1}} = \frac{-\ket{001} + \ket{010} + \ket{100}}{\sqrt{3}} $$


$$\ket{\phi_{2}} = \frac{\ket{001} - \ket{010} + \ket{100}}{\sqrt{3}} $$


$$\ket{\phi_{3}} = \frac{\ket{001} + \ket{010} - \ket{100}}{\sqrt{3}} $$

where the door numbers are encoded as:

 $\ket{001} \equiv \ket{0}$, $\ket{010} \equiv \ket{1}$, and $\ket{100} \equiv \ket{2}$,

Another Hint: You have access to all unitary quantum gates

Are your chances of winning more likely in Classical Monty Hall or in the Quantum Monty Hall problem?

### ***Solution***

Since we don't know whether or not we got $\ket{\phi_{1}}$, $\ket{\phi_{2}}$, or $\ket{\phi_{3}}$ as our wavefunction, we can write our receieved wavefunction from Monty arbitrarily as 

$$\ket{\phi} = \frac{-\ket{M} + \ket{d_{1}} + \ket{d_{2}} }{\sqrt{3}}$$

where we represent the $\ket{M}$ as the marked state with a minus sign (or we could think of 'M' for Monty's prize door) and $\ket{d_{1}}$ and $\ket{d_{2}}$ are both dud doors.


Luckily, we have a quantum algorithm that can help us detect states with minus signs -- Grover's algorithm! Grover's algorithm utilizes amplitude amplification in order to amplify the probability of measurement for a state marked by a minus sign.

We can think of each complex amplitude having its own spot on the complex plane



We can represent the mean point on this graph as the "center of mass" of these two points. Let's call the mean point $\alpha$.

$$\alpha = \frac{(-\frac{1}{\sqrt{3}})(1) + (2)(\frac{1}{\sqrt{3}})}{3} = \frac{1}{3\sqrt{3}}$$





Next we prepare to do reflection about the mean point by drawing reflected distances. Reflection about the mean point is equivalent to amplitude amplification in Grover's algorithm








Therefore, our final state of the wavefunction after amplitude amplification is a superposition state of the marked state $\ket{M}$ with a complex amplitude $\frac{5}{3\sqrt{3}}$ and two dud states $\ket{d_{1}}$ and $\ket{d_{2}}$ each with complex amplitude $-\frac{1}{3\sqrt{3}}$.

Therefore, the transformed wavefunction after amplitude amplification is

$$\ket{\psi'} = \frac{5}{3\sqrt{3}} \ket{M} - \frac{1}{3\sqrt{3}} \ket{d_{1}} - \frac{1}{3\sqrt{3}} \ket{d_{2}} $$

The probability of measuring $\ket{M}$ is $(\frac{5}{3\sqrt{3}})^{2} = \frac{25}{27} \approx 0.93$.

Although Monty believes that you have a 1/3 chance of guessing correctly based on his randomly assigned minus sign, you actually have a $93 \% $ chance of guessing correctly if you apply the amplitude amplification unitaries onto Monty's wavefunction that he gives you.

### ***Solution***

You choose door 0 since that was the state that you measured when you performed amplitude amplification. Assuming that the measured state was most likely the marked state, from what we found out in the previous part, Door 0 has a $93 \%$ chance of winning, door 1 has a $3.5 \%$ chance of winning, and door 2 has a $3.5 \%$ chance of winning. Since Monty revealed that door 1 was a dud, our probabilities now update in a Bayesian way. Door 2 absorbs the probabilities that Door 1 once had, and now after Monty's revealation, Door 2 has a $7 \%$ chance of winning. However, staying with the $93 \%$ chance of success that door 0 has to offer is a much better deal! We do not switch doors in the Quantum Monty Hall.
